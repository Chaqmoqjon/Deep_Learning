```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
import numpy as np
```

##Classik dasturlash
Kodni o'zimiz yozdik

```python
#data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=37)
x_data, y_data = data[0], data[1]

#custom func
def classic_program(x): return x*70
y_pred = [classic_program(x) for x in x_data]

#plotting
plt.plot(x_data, y_pred, label='pred', c='r')
plt.scatter(x_data, y_data, label='asl', c='g')
plt.title('Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

##Weight qo'shish
Funksiyani o'zgartirmasdan dataga og'irlik qo'shib ma'lumot aniqligini oshirdik

```python
#data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=37)
x_data, y_data = data[0], data[1]

#custom func
def weight_assigned_program(x, w): return w*x*70

w = 1.3

y_pred = [weight_assigned_program(x, w) for x in x_data]


#plotting
plt.plot(x_data, y_pred, label='pred', c='r')
plt.scatter(x_data, y_data, label='vaznli dastur', c='g')
plt.title('Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

##LinearRegression
Chiziqni o'zi eng aniq o'lchamda chizdi. ML ning o'zi eng optimal vazn (weight) berdi

```python
#data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=37)
x_data, y_data = data[0], data[1]

#regression
regression = LinearRegression().fit(x_data, y_data)
y_pred = regression.predict(x_data)

#plotting
plt.plot(x_data, y_pred, label='ML dastur', c='r') #chiziqning kordinatalari
plt.scatter(x_data, y_data, label='Asl Data', c='g') #nuqtalarning kordinatalari
plt.xlabel('x')
plt.ylabel('y')
equetion = f"{regression.coef_[0]:.3f}  x+{regression.intercept_:.3f} "
plt.title(equetion)
plt.legend()
plt.show()
```

##MLPRegressor

```python
#data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=37)
x_data, y_data = data[0], data[1]

#regression
regression = MLPRegressor(max_iter=2000).fit(x_data, y_data)
y_pred = regression.predict(x_data)

#plotting
plt.plot(x_data, y_pred, label='DL dastur', c='r') #chiziqning kordinatalari
plt.scatter(x_data, y_data, label='Asl Data', c='g') #nuqtalarning kordinatalari
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
